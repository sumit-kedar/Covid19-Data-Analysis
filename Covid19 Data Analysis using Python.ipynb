# Covid19 Data Analysis

```python
# Importing modules
```

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Importing Dataset
```

```python
corona_dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset.head(10)
```

```python
# Shape of dataframe
```

```python
corona_dataset.shape
```

```python
# Deleting useless columns
  # As Latitude and longitude columns are of no use
```

```python
corona_dataset.drop(["Lat","Long"],axis=1,inplace= True)
```

```python
corona_dataset.head(10)
```

```python
# Aggregating the rows by the country
```

```python
corona_dataset_aggregated = corona_dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape
```

# Visualizing data related to a country for example Italy

```python
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["US"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["Italy"].plot()
```

```python
# Calculating first derivative of the curve
```

```python
corona_dataset_aggregated.loc["Italy"].diff().plot()
```

```python
# find maxmimum infection rate for Italy
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
# find maximum infection rate for all of the countries
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries :
    max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
```

```python
# Adding Max infection rate column in dataset
```

```python
corona_dataset_aggregated['max infection rate'] = max_infection_rates
```

```python
corona_dataset_aggregated.head(4)
```

```python
# create a new dataframe with only needed column(only country names and max infection rate)
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max infection rate"])
```

```python
corona_data.head(5)
```

```python
# Importing new data for worldwide happiness report
```

```python
worldwide_happiness_report = pd.read_csv("worldwide_happiness_report.csv")
worldwide_happiness_report.head()
```

```python
worldwide_happiness_report.shape
```

```python
# Deleting useless columns
```

```python
columns_to_dropped = ["Overall rank","Score","Generosity","Perceptions of corruption"]
worldwide_happiness_report.drop(columns_to_dropped,axis=1 , inplace=True)
```

```python
worldwide_happiness_report.head()
```

```python
worldwide_happiness_report.set_index(['Country or region'],inplace=True)
worldwide_happiness_report.head()
```

```python
# joining two datasets
```

```python
data = worldwide_happiness_report.join(corona_data).copy()
data.head()
```
